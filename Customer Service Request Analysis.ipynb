```python

data.sort_values(["Complaint Type","Request Closing Time"],ascending=[False,True],inplace=True)
data
```

```python
data["Request Closing Time"] = pd.to_numeric(data["Request Closing Time"])
data
groups = data.groupby(["Location Type","Complaint Type"]).agg(Avg_Request_Closing_Time = ("Request Closing Time","mean"))
groups.sort_values("Avg_Request_Closing_Time",ascending=False)
```

alpha = 0.05
Null = Mean response time for each complaint type is similar
Alternate = Mean response time foreach complaint type is not similar
If p < 0.05, reject Null
If p > 0.05, fail to reject the null

```python
import scipy.stats as stats
a = groups["Avg_Request_Closing_Time"].to_numpy()
b = stats.ttest_1samp(a,71)
if b.pvalue <= 0.05:
    print("We reject the null hypothesis and conclude there is a difference in the average response time for each complaint type")
else:
    print("We fail to reject the null hypothesis and conclude there is no difference in the average response time for each complaint type")
```

Based off this test we can also conclude that the complaint type and the location type are related
